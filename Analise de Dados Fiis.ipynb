```python
import bs4
import urllib.request as urllib_request
import pandas as pd

print("BeautifulSoup ->", bs4.__version__)
print("urllib ->", urllib_request.__version__)
print("pandas ->", pd.__version__)
```

```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError

url = 'https://www.fundamentus.com.br/fii_resultado.php'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}

try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    html = response.read()
    print(html)
    
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e:
    print(e.reason)
 
```

```python
def trata_html(input):
    return " ".join(input.split()).replace('> <', '><')
```

```python
html = html.decode('ISO-8859-1')
html = trata_html(html)
html
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
dict_FII = {}
```

```python
def retorna_keys(soup):
    list_keys = []
    soup.findAll('th')
    for item in soup.findAll('th'):
        list_keys.append((item.get_text()))
    return list_keys

def retorna_codigo(soup):
    list_codigob3 = []
    for item in soup.findAll('td', {"ckass":"res_papel"}):
        if(item.get_text().find("11") == 4):
            list_codigob3.append(item.get_text())
    return list_codigob3

def retorna_segmento(soup):
    list_segmento = []
    for item in soup.findAll('td', {"ckass":"res_papel"}):
        if(item.get_text().find("11") != 4):
            list_segmento.append(item.get_text())
    return list_segmento
```

```python
list_keys = retorna_keys(soup)
print(len(list_keys))
```

```python
list_codigob3 = retorna_codigo(soup)
```

```python
list_segmento = retorna_segmento(soup)
```

```python
soup.findAll('table')
```

```python
table = soup.findAll('table')
str_table = str(table)
str_table = str_table.replace("%","")
str_table = str_table.replace(".","")
df_fiis = pd.read_html(str_table,decimal=',')[0]
df_fiis
```

```python
list_endereco = []
soup.findAll('td', {'class':'endereco'})
for item in soup.findAll('td', {'class':'endereco'}):
    list_endereco.append(item.get_text())
print(len(list_endereco))
```

```python
df_fiis.insert(13,"Endereço",pd.Series(list_endereco))
display(df_fiis)
```

```python
df_fiis['Cotação'] = df_fiis['Cotação']/100
df_fiis['FFO Yield'] = df_fiis['FFO Yield']/100
df_fiis['Dividend Yield'] = df_fiis['Dividend Yield']/100
df_fiis['P/VP'] = df_fiis['P/VP']/100
df_fiis['Cap Rate'] = df_fiis['Cap Rate']/100
df_fiis['Vacância Média'] = df_fiis['Vacância Média']/100
```

```python
import plotly.express as px
```

### Lajes Corporativas

```python
df_fiis_lajes = df_fiis.loc[df_fiis["Segmento"] == "Lajes Corporativas"].sort_values(by="P/VP",ascending=False)
df_fiis_lajes = df_fiis_lajes.loc[df_fiis_lajes["P/VP"] < 2]
df_fiis_lajes = df_fiis_lajes.loc[df_fiis_lajes["Cotação"] < 250].reset_index()
df_fiis_lajes
```

```python
fig = px.histogram(df_fiis_lajes, x="P/VP",nbins=15)
fig.show()
fig = px.box(df_fiis_lajes, y="P/VP")
fig.show()
```

### Shoppings

```python
df_fiis_shoppings = df_fiis.loc[df_fiis["Segmento"] == "Shoppings"].sort_values(by="P/VP",ascending=False)
df_fiis_shoppings = df_fiis_shoppings.loc[df_fiis_shoppings["Cotação"] < 200]
df_fiis_shoppings = df_fiis_shoppings.loc[df_fiis_shoppings["Dividend Yield"] > 4]
df_fiis_shoppings = df_fiis_shoppings.loc[df_fiis_shoppings["Liquidez"] > 0].reset_index()
df_fiis_shoppings
```

```python
fig = px.histogram(df_fiis_shoppings, x="Dividend Yield",nbins=20)
fig.show()
fig = px.box(df_fiis_shoppings, y="Dividend Yield")
fig.show()
```

### Galpões

```python
df_fiis_galpoes = df_fiis.loc[df_fiis["Segmento"] == "Logística"].sort_values(by="P/VP",ascending=False)
df_fiis_galpoes = df_fiis_galpoes.loc[df_fiis_galpoes["Qtd de imóveis"] > 1].reset_index()
print(df_fiis_galpoes.shape)
df_fiis_galpoes
```

### Papéis

```python
df_fiis_papeis = df_fiis.loc[df_fiis["Segmento"] == "Títulos e Val Mob"].sort_values(by="Dividend Yield",ascending=False).reset_index()
df_fiis_papeis
```

### Fundos de Fundos

```python
df_fiis_fof = df_fiis.loc[df_fiis["Segmento"] == "Híbrido"].sort_values(by="Dividend Yield",ascending=False)
df_fiis_fof = df_fiis_fof.loc[df_fiis_fof["Liquidez"] > 0]
df_fiis_fof = df_fiis_fof.loc[df_fiis_fof["P/VP"] < 1].reset_index()
df_fiis_fof
```
