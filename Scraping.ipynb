```python
import bs4
import urllib.request as urllib_request
import pandas as pd

print("BeautifulSoup ->", bs4.__version__)
print("urllib ->", urllib_request.__version__)
print("pandas ->", pd.__version__)
```

```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError

url = 'https://www.fundamentus.com.br/fii_resultado.php'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}

try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    html = response.read()
    print(html)
    
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e:
    print(e.reason)
 
```

```python
def trata_html(input):
    return " ".join(input.split()).replace('> <', '><')
```

```python
html = html.decode('ISO-8859-1')
html = trata_html(html)
html
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
def retorna_keys(soup):
    list_keys = []
    soup.findAll('th')
    for item in soup.findAll('th'):
        list_keys.append((item.get_text()))
    return list_keys

def retorna_codigo(soup):
    list_codigob3 = []
    for item in soup.findAll('td', {"ckass":"res_papel"}):
        if(item.get_text().find("11") == 4):
            list_codigob3.append(item.get_text())
    return list_codigob3

def retorna_segmento(soup):
    list_segmento = []
    for item in soup.findAll('td', {"ckass":"res_papel"}):
        if(item.get_text().find("11") != 4):
            list_segmento.append(item.get_text())
    return list_segmento
```

```python
list_keys = retorna_keys(soup)
print(len(list_keys))
```

```python
list_codigob3 = retorna_codigo(soup)
```

```python
list_segmento = retorna_segmento(soup)
```

# Filtro por Segmento
## Segmentos escolhidos
    - Lajes
    - Shoppings
    - Galpes
    - Papis
    - Fundos de Fundos
   

### Lajes Corporativas

```python
df_fiis_lajes = df_fiis.loc[df_fiis["Segmento"] == "Lajes Corporativas"].sort_values(by="P/VP",ascending=False).reset_index()
df_fiis_lajes
```

### Shoppings

```python
df_fiis_shoppings = df_fiis.loc[df_fiis["Segmento"] == "Shoppings"].sort_values(by="Dividend Yield",ascending=False).reset_index()
df_fiis_shoppings
```

### Galpes

```python
df_fiis_galpoes = df_fiis.loc[df_fiis["Segmento"] == "Logstica"].sort_values(by="Dividend Yield",ascending=False).reset_index()
df_fiis_galpoes
```

### Papis

```python
df_fiis_papeis = df_fiis.loc[df_fiis["Segmento"] == "Ttulos e Val. Mob."].sort_values(by="Dividend Yield",ascending=False).reset_index()
df_fiis_papeis = df_fiis_papeis.loc[df_fiis_papeis['P/VP'] < 100]
df_fiis_papeis
```

### Fundos de Fundos

```python
df_fiis_fof = df_fiis.loc[df_fiis["Segmento"] == "Hbrido"].sort_values(by="Dividend Yield",ascending=False).reset_index()
df_fiis_fof = df_fiis.loc[df_fiis["P/VP"] < 100]
df_fiis_fof
```
